```python
import pandas as pd
```

```python
sales_df = pd.read_csv('Sales.csv')
customers_df = pd.read_csv('Customers.csv')
products_df = pd.read_csv('Products.csv')
store_df = pd.read_csv('Stores.csv')
exchange_rates_df = pd.read_csv('Exchange_rates.csv')
```

```python
merged_df = pd.merge(sales_df, customers_df, on='customerkey', how='left')
merged_df = pd.merge(merged_df, products_df, on='productkey', how='left')
merged_df = pd.merge(merged_df, store_df, on='storekey', how='left')
#merged_df = pd.merge(merged_df, exchange_rates_df, left_on='order_date', right_on='date', how='left')
#merged_df = pd.merge(merged_df, exchange_rates_df, left_on='currency_code', right_on='currency_code', how='left')
merged_df = pd.merge(merged_df, exchange_rates_df, left_on=['order_date', 'currency_code'], right_on=['date', 'currency_code'], how='left')
```

```python
merged_df.columns = merged_df.columns.str.lower()
merged_df.columns = merged_df.columns.str.replace(' ','_')
```
